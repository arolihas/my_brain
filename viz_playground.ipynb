**Brain Images**

```python
path = os.getcwd() + '/nifti'
```

```python
files = os.listdir(path)
```

```python
image_names = [path + '/' + img for img in files if img.endswith('.nii')]
```

```python
four_d = []
for i, img in enumerate(image_names):
    try:
        print(img)
        plotting.plot_img(img)
        plt.show()
    except:
        print('4D')
```
